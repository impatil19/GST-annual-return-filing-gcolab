```python
# reading excel file from google drive
import os
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Vinayaka.xlsx')
```

```python
df.head(2)
```

```python
df['GSTIN of supplier'].value_counts()
```

```python
# select only the numeric columns to sum
numeric_cols = df.select_dtypes(include='number').columns
numeric_cols = numeric_cols.drop('Rate (%)')

# group by the 'GSTIN of supplier' and 'Rate (%)' columns, and sum the numeric columns
result = df.groupby(['GSTIN of supplier', 'Rate (%)'], as_index=False)[numeric_cols].sum()
```

```python
result.head(2)
```

```python
# storing the output file in google drive, change the filename at the end which stores file in gdrive
result.to_excel(excel_writer=r'/content/drive/MyDrive/Colab Notebooks/Vinayanka_agg.xlsx')
```
